```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# checking if there is any missing values

df.isnull().sum()
```

```python
# x ke basis pr y determine hoga
# y jo h woh x pr depend kr raha h
```

```python
# regression - unlimited data (infinite)
# classification - limited classes
df['owner'].value_counts()
```

```python
dt = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fourth Owner Or More':4}
dt
```

```python
#  for ownership in df['owner']:
#       print(dt[ownership])

df['owner'] = df['owner'].map(dt)
df
```

```python
df['brand'].nunique()

# total bike brands
```

```python
df['brand'].value_counts()
```

```python
brand_ls = list(df['brand'].value_counts().head(16).keys())
brand_ls
```

```python
df2 = df[df['brand'].isin(brand_ls)]
df2.head()
```

```python
brand_dic={}
cnt=1
for i in list(df2['brand'].value_counts().keys()):
    brand_dic[i] = cnt
    cnt+=1

brand_dic
```

```python
df2['brand'] = df2['brand'].map(brand_dic)
```

```python
df2.head()
```

```python
df2.drop(['bike_name','city'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
x = df2.drop('price',axis=1)
y = df2[['price']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# scikit-learn
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
# 80% training size
```

```python
from sklearn.linear_model import LinearRegression 
# algorithm import
```

```python
lnr = LinearRegression()
```

```python
lnr
```

```python
lnr.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
x_train.isnull().sum()
```

```python
x_train
```

```python
print('Training score',lnr.score(x_train,y_train)*100)
```

```python
print('Testing score',lnr.score(x_test,y_test)*100)
# unseen data or test data or validation data (x_test,y_test)
```

```python
# prediction
prediction = lnr.predict(x_test)
```

```python
y_test
# original bike prices
```

```python
y_test['prediction'] = prediction 
```
